##2. Background music with audio

```python
# Install
!pip install gTTS pydub -q
!apt-get install -y -qq ffmpeg

# Generate short narration with gTTS
from gtts import gTTS

text = "Under the moonlight, the river sings softly with fireflies dancing above."
tts = gTTS(text, lang="en")
tts.save("voice.mp3")

# Create a soft ambient background tone
from pydub.generators import Sine
from pydub import AudioSegment

# 15-second pad at 220 Hz, reduced volume
pad = Sine(220).to_audio_segment(duration=15000).apply_gain(-10)

# Load generated voice
voice = AudioSegment.from_file("voice.mp3").set_frame_rate(44100).set_channels(2)

# Mix voice + ambient pad, start voice after 1 second
combined = pad.overlay(voice, position=1000)

# Export combined audio
combined.export("simple_music.wav", format="wav")

# Play audio in Colab / Jupyter
from IPython.display import Audio
Audio("simple_music.wav")
```
